```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
car_data = pd.read_csv('Cars93.csv')

# Inline indicates to present graphs as cell output
# read_csv returns a DataFrame, the file path is relative to that of your notebook.
```

##### Using Group-byâ€™s and Merges
Group-byâ€™s can be used to build groups of rows based on a specific feature in your dataset eg. the â€˜Manufacturerâ€™ categorical column. 

```python
group_by_Manufacturer = car_data.groupby(by=['Manufacturer'])

car_data_avg = round(group_by_Manufacturer.mean(),0)
car_data_avg
```

##### With the generated dataset above we are able to do better visualizations for the data set 

```python
# Since all the columns in car_data_count are the same, we will use just the first column as the rest yield
#the same result. iloc allows us to take all the rows and the zeroth column.

car_data_count_series = car_data_avg.iloc[5:5,1]


## we will assemble a DataFrame of only the relevant features to plot  
features_of_interest = pd.DataFrame({'EngineSize': car_data_avg['EngineSize'],
                                     'Passengers': car_data_avg['Passengers'],
                                   }
                                   )

## Let us plot a few of the manufactures for visibility
features_of_interest = features_of_interest.iloc[:10,]
features_of_interest.plot(kind='bar')




```

#### Want to try pie charts? Lets do it

##### Lets prepare some data. We can create a pie chart to show the population of categorical variables 

```python
# group by passengers 
group_by_Passengers = car_data.groupby(by=['Type'])

car_typePassengers = group_by_Passengers.count()

# get first column
car_typePassengers = pd.DataFrame(car_typePassengers.iloc[:,0])

car_typePassengers.rename(columns={'Manufacturer': 'Passengers'}, inplace=True)

car_typePassengers.plot.pie(subplots=True,figsize=(8, 8))

```

```python



# group by Type 
group_by_Type = car_data.groupby(by=['Type'])

car_typecount = group_by_Type.count()

car_typecount = pd.DataFrame(car_typecount.iloc[:,0])

car_typecount.rename(columns={'Manufacturer': 'Type'}, inplace=True)

car_typecount.plot.pie(subplots=True,figsize=(8, 8))
```

```python
# Group by Airbags
group_by_AirBags = car_data.groupby(by=['AirBags'])

car_AirBagscount = group_by_AirBags.count()

car_AirBagscount = pd.DataFrame(car_AirBagscount.iloc[:,0])

car_AirBagscount.rename(columns={'Manufacturer': 'Airbags'}, inplace=True)


car_AirBagscount.plot.pie(subplots=True,figsize=(8, 8))
```

```python
plt.figure(figsize = (15,15))
x = car_data['MPG.city']
y = car_data['MPG.highway']
plt.scatter(x,y, lw = (20), color = "g", facecolor = 'r')
plt.xlabel("MPG City", fontsize = (15))
plt.ylabel("MPG Higway", fontsize = (15))
plt.title("MPG City vs Highway", fontsize = (25))

plt.show()
```

```python

corr = car_data.corr()
plt.figure(figsize=(20,20))
plt.title("Heatmap of all Data", fontsize = (40))
plt.imshow(corr,cmap = 'hot')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=20, fontsize =(12))
plt.yticks(range(len(corr)), corr.columns , fontsize = (20))
plt.show()
```

```python
data = [car_data['MPG.city'], car_data['MPG.highway']]
plt.violinplot(data,showmeans = True)
plt.ylabel("Miles")
plt.title("MPG City vs Highway")
plt.show()
```

```python
plt.title("Difference in car Prices ")
data = [car_data["Min.Price"], car_data["Price"], car_data["Max.Price"]]
plt.boxplot(data)
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.title("Prices according to vehicle Type", fontsize =(25))
x = car_data["Type"]
y = car_data["Price"]
plt.ylabel("Price", fontsize = (15))
plt.xlabel("Vehicle Type", fontsize = (15))
plt.scatter(x,y, marker = "D", lw = (10), color = "orange", facecolor = 'white')
plt.show()
```

```python
plt.title("Car Makers Origin")
plt.plot(car_data["EngineSize"], label = "Enginesize")
plt.plot(car_data["Horsepower"], label = "Horsepower")
plt.plot(car_data["RPM"] / 100, label = "Rpm / 100")


plt.legend()
plt.show()
```

```python
plt.title("Miles per Tank, City vs Highway")
miles_per_tank_city = car_data["Fuel.tank.capacity"] * car_data["MPG.city"]
miles_per_tank_hw = car_data["Fuel.tank.capacity"] * car_data["MPG.highway"]

plt.plot(miles_per_tank_city, label = "City", ls = (":"), color = "r")
plt.plot(miles_per_tank_hw, label = "Highway", ls = (":"), color = "b")
plt.legend()
plt.show()
```
